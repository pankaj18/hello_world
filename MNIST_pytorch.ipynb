```python
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
from torchvision import datasets,transforms
from tqdm.notebook import tqdm
```

```python
train_data=torchvision.datasets.MNIST('./download',download=True,train=True,transform=transforms.ToTensor())
test_data=torchvision.datasets.MNIST('./downnload',download=True,train=False,transform=transforms.ToTensor())
```

```python
train_loader=torch.utils.data.DataLoader(train_data,batch_size=128,shuffle=True)
test_loader=torch.utils.data.DataLoader(test_data,batch_size=128,shuffle=False)
```

```python
class MNIST_CNN(nn.Module):
  def __init__(self):
    super().__init__()
    self.conv1=nn.Conv2d(1,32,3)
    self.conv2=nn.Conv2d(32,32,3)
    self.conv3=nn.Conv2d(32,64,3)
    self.conv4=nn.Conv2d(64,64,3)
    self.fc1=nn.Linear(64*4*4,256)
    self.fc2=nn.Linear(256,10)
    self.maxpool2d=nn.MaxPool2d(kernel_size=2)
  def forward(self,x):
    x=F.relu(self.conv1(x))
    x=F.relu(self.conv2(x))
    x=self.maxpool2d(x)
    x=F.relu(self.conv3(x))
    x=F.relu(self.conv4(x))
    x=self.maxpool2d(x)
    #print(x.shape)
    x=x.view(-1,64*4*4)
    x=F.relu(self.fc1(x))
    x=self.fc2(x)
    return x
```

```python
model=MNIST_CNN()
device=torch.device("cuda:0")
model=model.to(device)
```

```python
optimizer=torch.optim.Adam(model.parameters(),lr=0.01)
```

```python

criterion=nn.CrossEntropyLoss()
```

```python
epochs=30
for epoch in range(epochs):
  for images,labels in tqdm(train_loader):
    images,labels=images.to(device),labels.to(device)
    images=images.view(-1,1,28,28)
    out=model(images)
    optimizer.zero_grad()
    loss=criterion(out,labels)
    loss.backward()
    optimizer.step()
```

```python
correct=0
total=len(test_data)
with torch.no_grad():
  for images,labels in tqdm(test_loader):
    images,labels=images.to(device),labels.to(device)
    images.view(-1,1,28,28)
    out=model(images)
    pred_labels=torch.argmax(out,dim=1)
    correct+=torch.sum((labels==pred_labels).float())

print("accuracy is {}".format(correct/total))
```
