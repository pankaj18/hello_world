```python
columns=list(data.iloc[0,1:17])
columns
```

```python
data_thia=data.iloc[1:5,1:17]
data_thia.columns=columns
data_thia.reset_index(drop=True)
```

```python
def calculate_errors(df):
  error_df=pd.DataFrame()
  for col in list(df.columns):
    error_df[col+'_error_m']=df[col]-df['CCSD']
  print(error_df)
  error_df_calc=pd.DataFrame()
  error_df_calc['mue']=error_df.abs().mean(axis=0).values
  #print("mue\n")
  #print(error_df_calc['mue'])
  error_df_calc['mse']=error_df.mean(axis=0).values
  #print('\n mse \n')
  #print( error_df_calc['mse'])
  error_df_calc['max_ue']=error_df.abs().max(axis=0).values
  #print('\n max_ue \n ')
  #print(error_df_calc['max_ue'])
  error_df_calc.index=list(df.columns)
  return error_df_calc
```

```python
error_df_thia=calculate_errors(data_thia)
print(error_df_thia)
plt.figure()
sns.set_context('paper', font_scale=1.6)
error_df_thia.iloc[0:-1].plot.bar(rot=0,figsize=(16,6))
#plt.xlabel('')
plt.ylabel('$E_{DFT}-E_{CCSD}$ (eV)')
#plt.xlabel(columns[:-1],rotation=45)
plt.xticks(rotation=45) 
#plt.tight_layout()
plt.savefig('thia.jpg',dpi=300,bbox_inches='tight')
```

```python
def plot_scatter(df,columns):
  plt.figure(figsize=(14,6))

  s=50 #cahnge size of marker
  plt.scatter(df[columns[0]],df[columns[1]],marker="s",s=s,label=columns[1])
  plt.scatter(df[columns[0]],df[columns[2]],marker="8",s=s,label=columns[2])
  plt.scatter(df[columns[0]],df[columns[3]],marker="p",s=s,label=columns[3])
  plt.scatter(df[columns[0]],df[columns[4]],marker="P",s=s,label=columns[4])
  plt.scatter(df[columns[0]],df[columns[5]],marker="*",s=s,label=columns[5])
  plt.scatter(df[columns[0]],df[columns[6]],marker="h",s=s,label=columns[6])
  plt.scatter(df[columns[0]],df[columns[7]],marker="+",s=s,label=columns[7])
  plt.scatter(df[columns[0]],df[columns[8]],marker="x",s=s,label=columns[8])
  plt.scatter(df[columns[0]],df[columns[9]],marker="d",s=s,label=columns[9])
  plt.scatter(df[columns[0]],df[columns[10]],marker="o",s=s,label=columns[10])
  plt.scatter(df[columns[0]],df[columns[11]],marker="v",s=s,label=columns[11])
  plt.scatter(df[columns[0]],df[columns[12]],marker="^",s=s,label=columns[12])
  plt.scatter(df[columns[0]],df[columns[13]],marker="<",s=s,label=columns[13])
  plt.scatter(df[columns[0]],df[columns[14]],marker=">",s=s,label=columns[14])
  plt.scatter(df[columns[0]],df[columns[15]],marker="D",s=s,label=columns[15])
  plt.scatter(df[columns[0]],df[columns[16]],marker="H",s=s,label=columns[16])
  plt.legend(bbox_to_anchor = ( 1.0, 1))
  plt.xticks(np.arange(1,11,step=1))
  plt.yticks(np.arange(0,14,step=2))
  plt.xlabel('number of rings (n)')
  plt.ylabel('HOMO-LUMO gap (eV)')
  sns.set_context('paper', font_scale=1.6)
```

```python
data_thia=data.iloc[1:11,0:17]
data_thia.columns=columns
data_thia.reset_index(drop=True)
```

```python
plot_scatter(data_thia,columns)
plt.savefig('data_thia_scatter.jpg',dpi=300,bbox_inches='tight')
```
