```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_excel('Homo_Lumo_gap.xlsx')
```

```python
columns=list(data.iloc[0,1:16])
```

```python
def plot_scatter(df,columns):
  plt.figure(figsize=(14,6))
  plt.scatter(df['number of rings'],df[columns[1]],marker="s",label=columns[1])
  plt.scatter(df['number of rings'],df[columns[2]],marker="8",label=columns[1])
  plt.scatter(df['number of rings'],df[columns[3]],marker="p",label=columns[1])
  plt.scatter(df['number of rings'],df[columns[4]],marker="P",label=columns[1])
  plt.scatter(df['number of rings'],df[columns[5]],marker="*",label=columns[1])
  plt.scatter(df['number of rings'],df[columns[6]],marker="h",label=columns[1])
  plt.scatter(df['number of rings'],df[columns[7]],marker="+",label=columns[1])
  plt.scatter(df['number of rings'],df[columns[8]],marker="x",label=columns[1])
  plt.scatter(df['number of rings'],df[columns[9]],marker="d",label=columns[1])
  plt.scatter(df['number of rings'],df[columns[10]],marker="o",label=columns[1])
  plt.scatter(df['number of rings'],df[columns[11]],marker="v",label=columns[1])
  plt.scatter(df['number of rings'],df[columns[12]],marker="^",label=columns[1])
  plt.scatter(df['number of rings'],df[columns[13]],marker="<",label=columns[1])
  plt.scatter(df['number of rings'],df[columns[14]],marker=">",label=columns[1])
  plt.legend(bbox_to_anchor = ( 1.0, 0.8))
  plt.xticks(np.arange(1,11,step=1))
  plt.xlabel('n (number of rings)')
  plt.ylabel('HOMO-LUMO gap (eV)')
```

```python
plot_scatter(data_thia,columns)
```

```python
plot_scatter(data_seleno,columns)
```

```python
plot_scatter(data_telluro,columns)
```
