```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x1=[1,0.875,0.75,0.625,0.5,0.375,0.25,0.125,0]
y1=[-1.35,-1.29,-1.24,-1.19,-1.12,-1.07,-1.01,-0.93,-0.40]
x2=[1,0.83,0.67,0.5,0.33,0.17,0]
y2=[-1.37,-1.29,-1.11,-1.15,-1.06,-0.94,-0.85]
```

```python
plt.figure(figsize=(15,6))
plt.plot(x1,y1,label='x1')
plt.plot(x2,y2,label='x2')
plt.legend()
```

```python
plt.figure(figsize=(15,6))
plt.plot(x1,y1,'*',label='x1')
plt.plot(x2,y2,"^",label='x2')
plt.legend()
```
